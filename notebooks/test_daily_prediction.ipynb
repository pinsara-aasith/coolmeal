```python
import pandas as pd
import random

# Sample meal plans with their calories and nutritional values
meal_data = [
    {"Meal": "Low-Calorie Breakfast", "Calories": 350, "Protein": 20, "Fat": 10, "Fiber": 4},
    {"Meal": "High-Protein Lunch", "Calories": 700, "Protein": 50, "Fat": 25, "Fiber": 8},
    {"Meal": "Vegan Dinner", "Calories": 600, "Protein": 20, "Fat": 15, "Fiber": 10},
    {"Meal": "Keto Snack", "Calories": 300, "Protein": 10, "Fat": 25, "Fiber": 3},
    {"Meal": "Balanced Meal", "Calories": 500, "Protein": 30, "Fat": 15, "Fiber": 7},
    {"Meal": "High-Calorie Breakfast", "Calories": 800, "Protein": 40, "Fat": 35, "Fiber": 6},
    {"Meal": "Low-Fat Lunch", "Calories": 450, "Protein": 25, "Fat": 10, "Fiber": 5},
    {"Meal": "High-Fiber Dinner", "Calories": 650, "Protein": 30, "Fat": 20, "Fiber": 12},
    {"Meal": "Weight Gain Plan", "Calories": 1200, "Protein": 70, "Fat": 50, "Fiber": 8},
    {"Meal": "Low-Carb Snack", "Calories": 200, "Protein": 15, "Fat": 10, "Fiber": 2},
    {"Meal": "Protein-Rich Meal", "Calories": 750, "Protein": 60, "Fat": 30, "Fiber": 9},
    {"Meal": "Gluten-Free Lunch", "Calories": 550, "Protein": 30, "Fat": 18, "Fiber": 7},
    {"Meal": "Diabetic-Friendly Dinner", "Calories": 500, "Protein": 25, "Fat": 12, "Fiber": 9},
    {"Meal": "Heart-Healthy Meal", "Calories": 600, "Protein": 35, "Fat": 15, "Fiber": 10},
    {"Meal": "Weight Loss Plan", "Calories": 400, "Protein": 30, "Fat": 8, "Fiber": 6},
]

# Generate 50 records combining 3 meals
records = []
for _ in range(50):
    selected_meals = random.sample(meal_data, 3)
    total_calories = sum(meal['Calories'] for meal in selected_meals)
    total_protein = sum(meal['Protein'] for meal in selected_meals)
    total_fat = sum(meal['Fat'] for meal in selected_meals)
    total_fiber = sum(meal['Fiber'] for meal in selected_meals)

    record = {
        "Day Menu": f"{selected_meals[0]['Meal']} + {selected_meals[1]['Meal']} + {selected_meals[2]['Meal']}",
        "Total Calories": total_calories,
        "Total Protein (g)": total_protein,
        "Total Fat (g)": total_fat,
        "Total Fiber (g)": total_fiber
    }
    records.append(record)

# Convert to DataFrame
df = pd.DataFrame(records)

# Save to CSV
file_path = 'daily_menus.csv'
df.to_csv(file_path, index=False)

print(f"CSV file saved to: {file_path}")
```

#### Using Fuzzy Logic Map user preference to Nutrial values

In this i use dummy data and dummy rules , we must be correct it later

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Define the fuzzy variables and their ranges
age = ctrl.Antecedent(np.arange(0, 101, 1), 'age')
weight = ctrl.Antecedent(np.arange(30, 201, 1), 'weight')
height = ctrl.Antecedent(np.arange(100, 251, 1), 'height')
protein = ctrl.Consequent(np.arange(0, 201, 1), 'protein')
fat = ctrl.Consequent(np.arange(0, 101, 1), 'fat')
fiber = ctrl.Consequent(np.arange(0, 51, 1), 'fiber')

# Define membership functions for age
age['young'] = fuzz.trapmf(age.universe, [0, 0, 20, 30])
age['middle_aged'] = fuzz.trapmf(age.universe, [25, 35, 50, 60])
age['senior'] = fuzz.trapmf(age.universe, [55, 65, 100, 100])

# Define membership functions for weight
weight['underweight'] = fuzz.trapmf(weight.universe, [30, 30, 50, 60])
weight['normal'] = fuzz.trapmf(weight.universe, [55, 65, 75, 85])
weight['overweight'] = fuzz.trapmf(weight.universe, [80, 90, 110, 120])
weight['obese'] = fuzz.trapmf(weight.universe, [115, 130, 200, 200])

# Define membership functions for height
height['short'] = fuzz.trapmf(height.universe, [100, 100, 140, 155])
height['average'] = fuzz.trapmf(height.universe, [150, 165, 175, 185])
height['tall'] = fuzz.trapmf(height.universe, [180, 190, 250, 250])

# Define membership functions for protein
protein['low'] = fuzz.trapmf(protein.universe, [0, 0, 50, 75])
protein['medium'] = fuzz.trapmf(protein.universe, [60, 80, 120, 150])
protein['high'] = fuzz.trapmf(protein.universe, [140, 170, 200, 200])

# Define membership functions for fat
fat['low'] = fuzz.trapmf(fat.universe, [0, 0, 20, 30])
fat['medium'] = fuzz.trapmf(fat.universe, [25, 35, 55, 70])
fat['high'] = fuzz.trapmf(fat.universe, [65, 80, 100, 100])

# Define membership functions for fiber
fiber['low'] = fuzz.trapmf(fiber.universe, [0, 0, 10, 15])
fiber['medium'] = fuzz.trapmf(fiber.universe, [12, 18, 25, 35])
fiber['high'] = fuzz.trapmf(fiber.universe, [30, 40, 50, 50])

# Define the fuzzy rules
rule1 = ctrl.Rule(age['young'] & weight['normal'] & height['average'],
                  (protein['medium'], fat['medium'], fiber['high']))
rule2 = ctrl.Rule(age['middle_aged'] & weight['overweight'],
                  (protein['low'], fat['low'], fiber['medium']))
rule3 = ctrl.Rule(age['senior'] & weight['obese'] & height['short'],
                  (protein['low'], fat['low'], fiber['high']))
rule4 = ctrl.Rule(age['young'] & height['tall'],
                  (protein['high'], fat['medium'], fiber['medium']))
# Default rule
default_rule = ctrl.Rule(age['young'] | age['middle_aged'] | age['senior'],
                         (protein['medium'], fat['medium'], fiber['medium']))

# Create the control system and simulation
food_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, default_rule])
food_simulation = ctrl.ControlSystemSimulation(food_ctrl)

# Get inputs from the user
user_input = {
    "age": int(input("Enter your age (0-100): ")),
    "weight": int(input("Enter your weight in kg (30-200): ")),
    "height": int(input("Enter your height in cm (100-250): "))
}

# Input the values into the simulation
food_simulation.input['age'] = user_input['age']
food_simulation.input['weight'] = user_input['weight']
food_simulation.input['height'] = user_input['height']

try:
    # Perform the simulation
    food_simulation.compute()

    # Output the recommendations
    print(f"\nRecommended protein intake: {food_simulation.output['protein']:.2f} g")
    print(f"Recommended fat intake: {food_simulation.output['fat']:.2f} g")
    print(f"Recommended fiber intake: {food_simulation.output['fiber']:.2f} g")
except ValueError as e:
    print("Error during computation: ", e)
    print("Using default values due to insufficient rule activation.")
    # Provide default values
    print(f"\nRecommended protein intake: 100 g")
    print(f"Recommended fat intake: 50 g")
    print(f"Recommended fiber intake: 25 g")
```

### Get Total Calories using BMR

```python
def calculate_bmr(weight, height, age, gender):
    if gender.lower() == 'male':
        return 10 * weight + 6.25 * height - 5 * age + 5
    elif gender.lower() == 'female':
        return 10 * weight + 6.25 * height - 5 * age - 161
    else:
        raise ValueError("Gender must be 'male' or 'female'")

def calculate_daily_calories(bmr, activity_level):
    activity_factors = {
        'sedentary': 1.2,
        'lightly active': 1.375,
        'moderately active': 1.55,
        'very active': 1.725,
        'extra active': 1.9
    }
    return bmr * activity_factors.get(activity_level.lower(), 1.2)

def main():
    weight = float(input("Enter your weight in kg: "))
    height = float(input("Enter your height in cm: "))
    age = int(input("Enter your age in years: "))
    gender = input("Enter your gender (male/female): ")
    activity_level = input("Enter your activity level (sedentary, lightly active, moderately active, very active, extra active): ")

    try:
        bmr = calculate_bmr(weight, height, age, gender)
        daily_calories = calculate_daily_calories(bmr, activity_level)
        print(f"Your BMR is: {bmr:.2f} calories/day")
        print(f"Your daily calorie needs are: {daily_calories:.2f} calories/day")
    except ValueError as e:
        print(e)
main()
```
